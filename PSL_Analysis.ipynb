## PSL Dataset from 2016-2022

```python
import csv
dataset = list(csv.reader(open('PSL-2016-2022.csv')))
```

```python
dataset
```

```python
result = dataset[0].index('result')
team1 = dataset[0].index('team_1')
team2 = dataset[0].index('team_2')
for i in dataset[1:]:
    if i[result] in i[team1]:
        i[result]=i[team1]
    elif i[result] in i[team2]:
        i[result]=i[team2]
        
    
```

##### Number of matches teams won in a season

```python
year = dataset[0].index('psl_year')
result = dataset[0].index('result')
match = dataset[0].index('match_number')
season = {}
counter =0
for i in dataset[1:]:
    index = i[year]
    result_index=i[result]
    match_num = int(i[match])
    if index not in season:
        counter=1
        season[index]={result_index:1}
    else:
        if result_index not in season[index].keys():
            season[index][result_index]=0
        else:
            if counter != match_num:
                season[index][result_index]+=1
                counter+=1
```

```python
season
```

##### Total runs scored in a match of each season

```python
year = dataset[0].index('psl_year')
match = dataset[0].index('match_number')
runs = dataset[0].index('runs')
season = {}

for i in dataset[1:]:
    index = i[year]
    match_num=i[match]
    runs_index = int(i[runs])
    if index not in season:
       
        season[index]={'match-'+match_num:runs_index}
    else:
        if 'match-'+match_num not in season[index].keys():
            season[index]['match-'+match_num]=runs_index
        else:
            season[index]['match-'+match_num]+=runs_index
           
```

```python
season
```

##### The total runs scored in each inning of a match for each season

```python
year_index= dataset[0].index('psl_year')
match_index = dataset[0].index('match_number')
runs_index = dataset[0].index('runs')
inning_index = dataset[0].index('inning')
season = {}

for i in dataset[1:]:
    year = i[year_index]
    match=i[match_index]
    runs = int(i[runs_index])
    inning=i[inning_index]
    if year not in season:
        season[year]={'match-'+match:{inning:runs}}
    else:
        if 'match-'+match in season[year].keys():
            if inning not in season[year]['match-'+match].keys():
                season[year]['match-'+match][inning]=runs
            else:
                season[year]['match-'+match][inning]+=runs
        else:
            season[year]['match-'+match]={inning:runs}
           
  
```
